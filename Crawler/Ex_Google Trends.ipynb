### 透過迴圈抓取近一月的熱門關鍵字
因為我們在前面發現只要改日期我們就能抓到資料，因此在這裡我先設定兩個日期分別是今天與前29天。接著就透過迴圈逐一的去解析資料成 DataFrame 並 append 在一起就可以了

```python
import re
import json
import requests
import pandas as pd
from bs4 import BeautifulSoup
import datetime

enddt = datetime.datetime.today()
startdt = enddt - datetime.timedelta(days=29) #時間差
df = []
for i in pd.date_range(start=datetime.datetime.strftime(startdt,'%Y%m%d'), \
                       end=datetime.datetime.strftime(enddt,'%Y%m%d'), freq='1D'):
      url = 'https://trends.google.com.tw/trends/api/dailytrends?hl=zh-TW&tz=-480&ed={}&geo=TW&ns=15'.\
                    format(datetime.datetime.strftime(i, '%Y%m%d'))
      print(url)
      resp = requests.get(url)
      ndf=[]
      ndf = pd.DataFrame(json.loads(re.sub(r'\)\]\}\',\n', '', resp.text))\   #re.sub(substitute)是正則中的替換
                         ['default']['trendingSearchesDays'][0]['trendingSearches'])
      ndf['date'] = datetime.datetime.strftime(i, '%Y-%m-%d')
      df.append(ndf)
df = pd.concat(df, ignore_index=True)
df['title'] = df['title'].apply(lambda x: x['query'])
df
```

```python
df.to_excel("google_trends_20200708.xlsx")
```
