```python
import requests
res = requests.get('https://pages.tmall.com/wow/a/act/20548/wupr?spm=a211oj.12109038.leftnav007.16.48ba63e6FHkv5j&wh_pid=industry-154004&acm=lb-zebra-408096-5009122.1003.4.4485178&scm=1003.4.lb-zebra-408096-5009122.OTHER_13_4485178')
```

```python
res
```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(res.text, 'lxml')
soup
```

```python
import json
jd = json.loads(soup.select_one('#local-page-info').text)
```

```python
jd.keys()
```

```python
import pandas
products = []
for k in jd['data']:
    if 'list' in jd['data'][k]:
        df  = pandas.DataFrame(jd['data'][k]['list'])
        products.append(df)
```

```python
products_all = pandas.concat(products)
```

```python
products_all.columns
```

```python
products_all = products_all.loc[products_all['itemTitle'].notnull(), ['itemUrl', 'itemTitle', 'itemDesc','itemMonthSoldCount','itemActPrice', 'itemTagPrice']]
```

```python
products_all.info()
```

```python
products_all['itemActPrice'] = products_all['itemActPrice'].astype(float)
```

```python
products_all['itemTagPrice'] = products_all['itemTagPrice'].astype(float)
```

```python
products_all['discount'] = products_all['itemTagPrice'] - products_all['itemActPrice']
```

```python
products_all.sort_values('discount', ascending=False)
```
