```python
import requests
res = requests.get('https://search.taobao.tw/catalog/?_keyori=ss&ajax=true&from=input&page=2&q=iphone')
```

```python
jd = res.json()
jd
```

```python
import pandas
df = pandas.DataFrame(jd['mods']['listItems'])  #檢視網站發現資訊被放在mod下的listItems
```

```python
df.head()
```

```python
df.sort_values('price', ascending=False).head()  #根據價錢排序商品
```

```python
#抓取更多頁
import requests
products = []
for i in range(1,6):
    res = requests.get(f'https://search.taobao.tw/catalog/?_keyori=ss&ajax=true&from=input&page={i}&q=iphone') #f {i}:python3.7後支援功能
    jd = res.json()
    df = pandas.DataFrame(jd['mods']['listItems'])
    products.append(df)
```

```python
df_all = pandas.concat(products)
```

```python
len(df_all) #共有200個商品資訊
```
