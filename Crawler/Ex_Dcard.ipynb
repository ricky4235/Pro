# 抓取一篇Dcard的文章
具體項目如下：編號 / 標題 / 引言 / 內容 / 發布時間 / 更新時間…等等

```python
# 載入使用的套件
import pandas as pd
import requests
from requests_html import HTML
import re


# 檢視資料結構
#從網址列中可以看到這篇文章的編號是232646236，將這篇文章的編號透過quest套件讀取，並檢視抓下來資料的結構
ID = '232646236'
url = 'https://www.dcard.tw/_api/posts/' + ID
# 透過request套件抓下這個網址的資料
requ = requests.get(url)
# 初步檢視抓到的資料結構
requ.json()  #查看程式為json格式
```

```python
# 將抓下來的資料轉為DataFrame
ID = '231030181'
url = url = 'https://www.dcard.tw/_api/posts/' + ID
requ = requests.get(url)
rejs = requ.json()
pd.DataFrame(
    data=
    [{'ID':rejs['id'],
      'title':rejs['title'],
      'content':rejs['content'],
      'excerpt':rejs['excerpt'],
      'createdAt':rejs['createdAt'],
      'updatedAt':rejs['updatedAt'],
      'commentCount':rejs['commentCount'],
      'forumName':rejs['forumName'],
      'forumAlias':rejs['forumAlias'],
      'gender':rejs['gender'],
      'likeCount':rejs['likeCount'],
      'reactions':rejs['reactions'],
      'topics':rejs['topics']}],
    columns=['ID','title','content','excerpt','createdAt','updatedAt','commentCount','forumName','forumAlias','gender','likeCount','reactions','topics'])
```

確認可以透過程式把文章爬下來之後，我們就來寫個簡單的Crawl函數，期望只需要輸入文章的ID後，就回傳爬下來的文章內容！

```python
# 撰寫簡單的函數，透過輸入文章ID，就輸出文章的資料
def Crawl(ID):
    link = 'https://www.dcard.tw/_api/posts/' + str(ID)
    requ = requests.get(link)
    rejs = requ.json()
    return(pd.DataFrame(
        data=
        [{'ID':rejs['id'],
          'title':rejs['title'],
          'content':rejs['content'],
          'excerpt':rejs['excerpt'],
          'createdAt':rejs['createdAt'],
          'updatedAt':rejs['updatedAt'],
          'commentCount':rejs['commentCount'],
          'forumName':rejs['forumName'],
          'forumAlias':rejs['forumAlias'],
          'gender':rejs['gender'],
          'likeCount':rejs['likeCount'],
          'reactions':rejs['reactions'],
          'topics':rejs['topics']}],
        columns=['ID','title','content','excerpt','createdAt','updatedAt','commentCount','forumName','forumAlias','gender','likeCount','reactions','topics']))
```

```python
# 嘗試使用撰寫出的函數，抓取編號231030181的文章
Crawl(231030181)
```

# 一次讀取100篇最熱門的文章
透過系統提供的api，一次抓取100篇熱門文章<br>
https://www.dcard.tw/_api/posts?popular=true&limit=100

以下簡單說明這個網址

* popular參數：若設定為true，表示按照熱門程度排序，若設定為false，則按照發布時間排序
* limit參數：限定在0-100的數值，表示要抓多少文章

```python
url = 'https://www.dcard.tw/_api/posts?popular=true&limit=100'
resq = requests.get(url)
rejs = resq.json()
df = pd.DataFrame()
for i in range(len(rejs)):
    df = df.append(Crawl(rejs[i]['id']),ignore_index=True)
print(df.shape)
df
```

# 爬超過100篇文章
由於API限制最多載入100篇文章，如果我們想要爬更多資料，可以透過before參數與迴圈進行!<br>
before參數後面是接文章的ID，讓我們可以抓取某篇文章之前的資料<br>
而透過迴圈，我們只需要把之前抓到最後一篇文章的ID放入before參數中，我們就可以抓到這篇文章的前100篇文章。

```python
# 透過迴圈讀取10*100篇文章，若需讀取更多資料，可以將range(10)中的數值提升
for j in range(2):
    last = str(int(df.tail(1).ID)) # 找出爬出資料的最後一筆ID，tail()和head()一樣，可以在括號內加上要顯示幾筆資料
    url = 'https://www.dcard.tw/_api/posts?popular=true&limit=100&before=' + last
    resq = requests.get(url)
    rejs = resq.json()
    for i in range(len(rejs)):
        df = df.append(Crawl(rejs[i]['id']), ignore_index=True)
print(df.shape)
df
```

```python
#將資料轉換為excel保存到桌面
df.to_excel('Dcard.xlsx')
```

# 今日十大熱門文章 (無法爬)

```python
import requests
import re  #Regular Expression
from bs4 import BeautifulSoup


def main():
    dcard_topic_page = 'https://www.dcard.tw/f'
    resp = requests.get(dcard_topic_page)
    soup = BeautifulSoup(resp.text, 'html.parser')

    # ^ means "start with"
    topic_entry_pattern = '^PostEntry_container_'
    topic_title_pattern = 'strong'
    find_top10_hot_topic_title(soup, topic_entry_pattern, topic_title_pattern)


def find_top10_hot_topic_title(soup, topic_entry_pattern, topic_title_pattern):
    top_ten_topic = soup.find_all('div', {'class': re.compile(topic_entry_pattern)})
    #re.compile:用正則表達式來查詢目標，它單獨使用就沒有任何意義，需要和findall(), search(), match(）搭配使用。
    i = 1
    for topic in top_ten_topic[:10]:
        print(str(i) + ': ' + topic.find(topic_title_pattern).text)
        i += 1


if __name__ == '__main__':
    main()
```
