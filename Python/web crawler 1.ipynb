```python
import requests
from bs4 import BeautifulSoup
```

```python
r = requests.get("https://www.ptt.cc/bbs/MobileComm/index.html")
print(r.text)
```

```python
soup = BeautifulSoup(r.text,"html.parser")  #將抓下來的資料用Beautifulsoup4轉為HTML的parser
```

```python
sel = soup.select("div.title a") #因為我想選取的是網頁裡的文章標題，所以soup.select中放的才是div.title a
```

```python
for s in sel:
    print(s["href"], s.text)  #最後寫一個迴圈將爬下來的文章標題印出來
```
