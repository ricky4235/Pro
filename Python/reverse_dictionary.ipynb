```
# Please write a function to reverse the following nested input value into output value
 
# Input:
input_value = {
  'hired': {
    'be': {
      'to': {
        'deserve': 'I'
      }
    }
  }
}
 
```

```
# Output:
output_value = {
  'I': {
    'deserve': {
      'to': {
         'be': 'hired'
      }
    }
  }
}
```

```
input_value
```

```
output_value
```

```
def reverse_dictionary(dic,cid):
  if type(dic)==str:
    return {dic:cid}
  for key in dic:
    if type(cid)==None:
      return reverse_dictionary(dic[key],key)
    else:
      return reverse_dictionary(dic[key],{key:cid})
```

```
cid = reverse_dictionary(input_value,"")
```
