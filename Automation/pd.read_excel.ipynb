```python
#建議使用英文路徑以及英文命名方式。

import pandas as pd
io = r'C:\Users\11004076\Documents\Python Scripts\4_Automation\20200324-進銷存-SF4_Vietnam_Silicom.xlsx'
```

```python
# 2、sheet_name，要讀取的工作表名稱
```

* 可以是整型數字、列表名，也可以是兩種組成的列表。
* 整型數字：目標sheet所在的位置，以0為起始，比如sheet_name = 1代表第2個工作表。

```python
data = pd.read_excel(io, sheet_name = 1)  
data.head()
```

```python
data = pd.read_excel(io, sheet_name = '進銷存(圖)')
data.head()
```

```python
data = pd.read_excel(io, sheet_name = [1, 2], nrows = 5)  #代表讀取兩個工作表，返回兩個工作表組成的字典
data
```

```python
# 3、header， 用哪一行作列名
```

```python
 預設為0 ，如果設定為[0,1]，則表示將前兩行作為多重索引。
```

```python
data = pd.read_excel(io, sheet_name = 1, header = 0)  # 第一行作為列名。
#data = pd.read_excel(io, sheet_name = 1, header = [0,1])  # 前兩行作為列名。

data.head()
```

```python
# 4、names， 自定義最終的列名
```

* 一般適用於Excel缺少列名，或者需要重新定義列名的情況。
* 注意：names的長度必須和Excel列長度一致，否則會報錯。

```python
data = pd.read_excel(io, sheet_name = 1, 
                     names = ['Year','Month','import','sale','stock'])
data.head()
```

```python
# 5、index_col， 用作索引的列
```

```python
可以是工作表列名稱，如index_col = '排名'；
可以是整型或整型列表，如index_col = 0 或 [0, 1]，如果選擇多個列，則返回多重索引。
```

```python
data = pd.read_excel(io, sheet_name = '進銷存(圖)', index_col = 'Year')
data
```

```python
data = pd.read_excel(io, sheet_name = '進銷存(圖)', index_col = [0, 1])
data
```

```python
# 6、usecols，需要讀取哪些欄位
```

```python
可以使用整型，從0開始，如[0,2,3]；
可以使用Excel傳統的列名“A”、“B”等字母，如“A：C, E” ="A, B, C, E"，注意兩邊都包括。
usecols 可避免讀取全量資料，而是以分析需求為導向選擇特定資料，可以大幅提高效率。
```

```python
data = pd.read_excel(io, sheet_name = '進銷存(圖)', usecols = [0, 1, 3])
data.head()
```

```python
data = pd.read_excel(io, sheet_name = '進銷存(圖)', usecols = 'A:C, E')
data.head()
```

```python
# 7、squeeze(擠壓)，當資料僅包含一列
```

```python
 當資料僅包含一列時，squeeze為True時，返回Series，反之返回DataFrame。
```

```python
data = pd.read_excel(io, sheet_name = 'sq', squeeze  = True)
data.head()
```

```python
data = pd.read_excel(io, sheet_name = 'sq', squeeze  = False)
data.head()
```

```python
# 8、converters ，強制規定列資料型別
```

```python
data = pd.read_excel(io, sheet_name = 1, converters = {'Year': str, '進貨量': str, '存貨量': str})
#data['Year'].dtype
data
```

```python
# 9、skiprows，跳過特定行
```

```python
skiprows= n， 跳過前n行； skiprows = [a, b, c]，跳過第a+1,b+1,c+1行（索引從0開始）；
使用skiprows 後，有可能首行（即列名）也會被跳過。
```

```python
data = pd.read_excel(io, sheet_name = '進銷存(圖)', skiprows = [1,2,3])
# 跳過第2,3,4行資料（索引從0開始，包括列名）
data.head()
```

```python
data = pd.read_excel(io, sheet_name = '進銷存(圖)', skiprows = 3)
data.head()
```

```python
# 10、nrows ，需要讀取的行數
```

```python
如果只想瞭解Excel的列名及概況，不必讀取全量資料，nrows會十分有用。 
```

```python
data = pd.read_excel(io, sheet_name = '進銷存(圖)', nrows = 10)
data
```

```python
data = pd.read_excel(io, sheet_name = '進銷存(圖)', nrows = 15)
data
```

```python
# 11、skipfooter ， 跳過末尾n行
```

```python
data = pd.read_excel(io , sheet_name = '進銷存(圖)', skipfooter = 9) #跳過末尾行數（索引從0從末行開始算）
data
```
