```python
import xlwings as xw
import pandas as pd
import numpy as np

wb = xw.Book(r'C:\Users\11004076\Documents\Python Scripts\4_Automation\20200324-進銷存-SF4_Vietnam_Silicom.xlsx')
sht = wb.sheets['raw']
raw = sht.range('A1').expand('table')
df = pd.DataFrame(raw.value)  #將Excel的值轉化為DataFrame
df.columns = df.loc[0] #將欄位名稱和第0列一樣
df.drop(0, axis = 0, inplace=True)  #再將多餘的第0列丟掉
df.head()
```

```python
df[  (df['Year'] == 2020) & (df['ISS'] == '進貨量')  ]
```

```python
df.pivot_table(    #前面可以加篩選(excel樞紐的左上)
               columns = ['ISS'],   #欄變數(excel樞紐的右上)
               index = ['LargeCat','MediumCat'],  #列變數(excel樞紐的左下)
               values = ['Qty'],
               aggfunc=np.sum,
               #margins=True,  #添加所有行/列小計/總計
               #margins_name='Total', #當margins=True時，更改欄位名稱，默認為All
               fill_value=0  #用"某值"填滿遺漏值
               #dropna=True  #丟棄NA值
                )
```

```python
### 在Python用matplotlib做圖
```

```python
import matplotlib.pyplot as plt

df_ = df[['Year','Month', 'Qty', 'ISS']].groupby(by = ['Year','Month','ISS'])['Qty'].sum()
df_.plot(kind = 'bar', x = 'Year', y = 'Qty', title = 'x', legend = False)
plt.legend()
plt.show()
```

```python
x1 = [1,2,3,4,5]
y1 = [2,6,8,7,9]
y2 = [3,5,6,1,2]
# 比起堆疊長條圖，分組長條圖稍微調整數據以達到效果
width = 0.4 #先設定每個長條的寬度(0.5的話會無縫隙連在一起，大於0.5會重疊)
#注意這裡的X軸座標為x1中的數據減去寬度的一半(0.8, 1.8, 2.8, 3.8, 4.8)起寬度0.4的Bar
plt.bar([i-width/2 for i in x1], y1, width=width, label='G1') 
#注意這裡的X軸座標為x1中的數據加上寬度的一半(1.2, 2.2, 3.2, 4.5, 5.2)起寬度0.4的Bar
plt.bar([i+width/2 for i in x1], y2, width=width, label='G2') 
plt.legend()
plt.show()
```

```python
### 在Excel中做圖
要做分群的圖在Excel還是比較快
```

```python
sht_p = wb.sheets['chart']

plot = sht_p.charts.add()  #新增空白圖
plot.top = 300  #與頂端距離
plot.name = '進銷存'
plot.set_source_data(sht_p.range('chart!$A$1:$E$16'))  #範圍對象  
#plot.set_source_data(sht_p.range(sht_p.range('A1').expand()))
plot.api[1].SeriesCollection(1).XValues = '=chart!$A$2:$B$16'  #設定X軸數列 
plot.chart_type = 'column_clustered'  #柱狀圖
plot.height = 250
plot.width = 750
plot.api[1].SetElement(2)  # 設置圖表上方的標題
plot.api[1].ChartTitle.Text = '進銷存'
plot.api[1].HasLegend = 0  #使用HasLegend屬性刪除圖例。
plot.api[1].Axes(1).TickLabelSpacing = 1  #圖表軸上的每個元素都被顯示
plot.api[1].Axes(2).HasTitle = True   #座標軸標題(1:X軸、2:Y軸)
sht_p.shapes.api('進銷存').Line.Visible = 0  ##將Line對象的Visible屬性設置為0來刪除邊線。
```

```python
表達式chart.api返回帶有兩個COM包裝器的元組。我不太確定為什麼有兩個COM包裝器，
但是您似乎需要第二個包裝器才能訪問圖表。因此，此處使用chart.api [1]。
```
