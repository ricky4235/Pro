# 新增(生成新變數)
生成新變數，搭配 apply() 與 lambda 函數將公式應用到每一個觀測值。<br>
apply函數會自動遍歷每一行DataFrame的數據，最後將所有結果組合成一個Series數據結構並返回。

```python
df['Country_abb'] = df['Country'].apply(lambda y: y[3:5])
df.head(10)
```

```python
df['Revenue_TWD'] = df['Revenue'].apply(lambda y: y*30.1759256)
df.head(10)
```

### assign函數
為DataFrame分配新row。<br>
返回一個新對象，該對象包含除新row之外的所有原始row。重新分配的現有列將被覆蓋。

```python
df = pd.DataFrame({ 'temp_c' : [ 17.0 , 25.0 ]}, 
    index=[ 'Portland' , 'Berkeley' ]) 
df 
```

```python
df.assign(temp_f=lambda x: x.temp_c * 9 / 5 + 32 ) 
```

```python
#或者，可以通過直接引用現有的Series或序列來實現相同的行為:
df.assign(temp_f=df[ 'temp_c' ] * 9 / 5 + 32 ) 
```

```python
#在Python 3.6+中，您可以在同一個賦值中創建多個列，其中一個列依賴於同一個賦值中定義的另一個列:
df.assign(temp_f= lambda x: x[ 'temp_c' ] * 9 / 5 + 32 , 
          temp_k= lambda x: (x[ 'temp_f' ] +   459.67 ) * 5 / 9 )
```

# 連接:merge函數
SQL中的Join

```python
df1 = pd.DataFrame({'key': ['A', 'B', 'C', 'D'],
                    'value': np.random.randn(4)})

df2 = pd.DataFrame({'key': ['B', 'D', 'D', 'E'],
                     'value': np.random.randn(4)})
```

```python
df1
```

```python
df2
```

### 內連接
內連接使用比較運算符根據每個表共有的column的值匹配兩個表中的row，在SQL中實現內連接使用INNER JOIN

```python
pd.merge(df1, df2, on='key')
```

```python
#merge()提供了一些參數，可以將一个DataFrame的column与一个DataFeame的索引連接在一起
indexed_df2 = df2.set_index('key')  #將key改為index
pd.merge(df1, indexed_df2, left_on='key', right_index=True)
```

### 左/右外聯接
在SQL中實現左/右外連接可以使用LEFT OUTER JOIN和RIGHT OUTER JOIN<br>
在pandas中實現同樣可以使用merge()並指定how關鍵字為left或者right即可

```python
pd.merge(df1, df2, on='key', how='left')
```

### 全連接
全連接返回左表和右表中的所有行，無論是否匹配，但並不是所有的數據庫都支持，比如mysql就不支持，在SQL中實現全連接可以使用FULL OUTER JOIN

```python
#在pandas中實現同樣可以使用merge()並指定how關鍵字為outer
pd.merge(df1, df2, on='key', how='outer')
```

# 合併
SQL中UNION操作用於合併兩個或多個SELECT語句的結果集，UNION與UNION ALL類似，但是UNION將刪除重複的行。示例代碼如下

```python
#在pandas中可以使用concat()實現UNION ALL
pd.concat([df1, df2])
```

```python
#上面是UNION ALL保留重複值，如果希望刪除可以使用  drop_duplicates()
pd.concat([df1, df2]).drop_duplicates()
```

# 聚合函數.agg()
DataFrame.agg(func, axis=0, *args, **kwargs)

```python
df = pd.DataFrame([[1, 2, 3],
                   [4, 5, 6],
                   [7, 8, 9],
                   [np.nan, np.nan, np.nan]],
                  columns=['A', 'B', 'C'])
df
```

```python
#在row上聚合这些函数
df.agg(['sum', 'min'])
```

```python
#每个列有不同的聚合
df.agg({'A' : ['sum', 'min'], 'B' : ['min', 'max']})
```

```python
#对column进行聚合
df.agg("mean", axis="columns")
```

# 小計
其實就是新增欄位

```python
dfr = pd.DataFrame(np.random.randn(4, 5), columns=['A', 'B', 'C', 'D', 'E'])
dfr
```

### column小計

```python
#法一
dfr['Col_sum'] = dfr.apply(lambda x: x.sum(), axis=1)
dfr
```

```python
#法二
dfr['Col_sum_2']=dfr['A']+dfr['B']+dfr['C']+dfr['D']+dfr['E']
dfr
```

### row小計

```python
#法一
dfr.loc['Row_sum'] = dfr.apply(lambda x: x.sum())
dfr
```

```python
#法二
dfr.loc['Row_sum_2']=dfr.loc[0]+dfr.loc[1]+dfr.loc[2]+dfr.loc[3]
dfr
```

# pivot_table : pandas的樞紐分析
幾乎包羅上面所有方式<br>
df[WHERE].pivot_table().sort()

```python
df[ (df['LargeCat']=='Web Cam') ].pivot_table(    #前面可以加篩選(excel樞紐的左上)
               columns = ['LargeCat', 'Year'],   #column變數(excel樞紐的右上)
               index = ['Department','Country'],  #row列變數(excel樞紐的左下)
               values = ['Revenue'],              #excel樞紐的右下 Σ值
               aggfunc=np.sum,                    #計算方式
               margins=True,             #添加所有行/列小計/總計
               margins_name='Total',     #更改總計欄位名稱(預設為All，當margins=True時才會顯示)
               fill_value=0,  #用"某值"填滿遺漏值
               dropna=True    #丟棄NA值
                ).sort_index(    axis=1, ascending=False )  #排序
```
