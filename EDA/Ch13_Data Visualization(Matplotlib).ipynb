# Pyplot
matplotlib.pyplot是使matplotlib像MATLAB一樣工作的命令樣式函數的集合。每個pyplot功能都會對圖形進行一些更改：例如，創建圖形，在圖形中創建繪圖區域，在繪圖區域中繪製一些線條，用標籤裝飾繪圖等。

# .plot函數
plt.plot(x, y, format_string, **kwargs): 
* x為x軸數據，可為列表或數組；y同理；
* format_string為控製曲線的格式字符串，
* **kwargs第二組或更多的（x, y, format_string）

format_string：由顏色字符、風格字符和標記字符組成(先後順序沒差)。
* 顏色字符：'b'藍色；'#008000'RGB某顏色；'0.8'灰度值字符串
* 風格字符：'-'實線；'–'破折線； '-.'點劃線； '：'虛線； ''''無線條
* 標記字符：'.'點標記'o'實心圈'v'倒三角'^'上三角

# pyplot文本顯示函數
* plt.xlabel()：對x軸增加文本標籤
* plt.ylabel()：同理
* plt.title(): 對圖形整體增加文本標籤
* plt.text(): 在任意位置增加文本
* plt. annotate(s, xy = arrow_crd, xytext = text_crd, arrowprops = dict)：在圖形中增加帶箭頭的註解。 <br>
s表示要註解的字符串是什麼，xy對應箭頭所在的位置，xytext對應文本所在位置，arrowprops定義顯示的屬性

# Plot的圖表函數
* plt.plot(x,y , fmt) ：繪製坐標圖
* plt.boxplot(data, notch, position):繪製箱形圖
* plt.bar(left, height, width, bottom) :繪製條形圖
* plt.barh( width, bottom, left, height) :繪製橫向條形圖
* plt.polar(theta, r) :繪製極坐標圖
* plt.pie(data, explode) :繪製餅圖
* plt.scatter(x, y) :繪製散點圖
* plt.hist(x, bings, normed) :繪製直方圖

```python
#Ch13_1_1
import matplotlib.pyplot as plt

data = [-1, -4.3, 15, 21, 31]
plt.plot(data)  # x軸是 0,1,2,3,4
#您可能想知道為什麼x軸的範圍是0-3，而y軸的範圍是1-4。如果為plot()命令提供單個列表或數組 ，則matplotlib假定它是y值的序列，
#並自動為您生成x值。由於python範圍從0開始，因此默認的x向量的長度與y相同，但從0開始。因此x數據為 [0,1,2,3]。
plt.show()
```

```python
#plot()是一個通用命令，它將接受任意數量的參數。例如，要繪製x與y的關係，可以發出以下命令：
plt.plot([1, 2, 3, 4], [1, 4, 9, 16])
plt.show()
```

### 格式化繪圖樣式
對於每對x，y參數，都有一個可選的第三個參數，它是表示圖的顏色和線條類型的格式字符串。格式字符串的字母和符號來自MATLAB，您將顏色字符串與線型字符串連接在一起。默認格式字符串是“ b-”，這是一條藍色實線。例如，要用紅色圓圈繪製以上內容，您將發出

```python
plt.plot([1, 2, 3, 4], [1, 4, 9, 16], 'ro')
plt.axis([0, 6, 0, 20])  #axis()獲取一個列表並指定軸範圍。[xmin, xmax, ymin, ymax]
plt.show()
```

### 資料加上點標籤
zip()函數用於將可迭代的對像作為參數，將對像中對應的元素打包成一個個元組，然後返回由這些元組組成的列表。<br>
如果各個迭代器的元素個數不一致，則返回列表長度與最短的對象相同，利用* 號操作符，可以將元組解壓為列表。

```python
import matplotlib.pyplot as plt

days = range(0, 22, 3)
celsius = [25.6, 23.2, 18.5, 28.3, 26.5, 30.5, 32.6, 33.1]

#資料加上點標籤
for a,b in zip(days,celsius):
    plt.text(a, b, b, ha='center', va= 'bottom',fontsize=11)
    #plt.text(x, y, 要標住的字串, fontdict=None, withdash=False, **kwargs)

plt.plot(days, celsius, '^-.y')
plt.show()
```

### 繪製多組數據
有多種方法可以繪製多組數據。
* 一組: plot([x], y, [fmt], *, data=None, **kwargs)
* 多組: plot([x], y, [fmt], [x2], y2, [fmt2], ..., **kwargs) <br>
    fmt : 格式字符串，例如'ro'表示紅色圓圈 <br>
    kwargs : 用於指定屬性，例如線標籤（用於自動圖例），線寬，抗鋸齒，標記面顏色

```python
import numpy as np

t = np.arange(0., 5., 0.2) #np.arange(起始, 最後, 間隔)

plt.plot(t, t, 'r--')   
plt.plot(t, t**2, 'bs') 
plt.plot(t, t**3, 'g^')
plt.show()
```

```python
#Ch13_2_1b : 堆疊長條圖
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("C:\\Users\\11004076\\Documents\\Python Scripts\\FT748\\Ch13\\HOU_players_stats_2017_18.csv")
df_grouped = df.groupby("Pos")
points = df_grouped["PTS/G"].mean()  #各位置平均得分
rebounds = df_grouped["TRB"].mean()  #各位置平均籃板
# 繪出堆疊長條圖
plt.bar([1, 2, 3, 4, 5], points, label="Points")
plt.bar([1, 2, 3, 4, 5], rebounds, label="Rebounds")
plt.xticks([1, 2, 3, 4, 5], points.index)
plt.legend()
plt.ylabel("Points and Rebounds")
plt.xlabel("Position")
plt.title("NBA Houston Rockets") 
plt.show()
```

```python
#Ch13_2_1c : 群組長條圖
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("C:\\Users\\11004076\\Documents\\Python Scripts\\FT748\\Ch13\\HOU_players_stats_2017_18.csv")
df_grouped = df.groupby("Pos")
points = df_grouped["PTS/G"].mean()
rebounds = df_grouped["TRB"].mean()
# 繪出群組長條圖
index = range(1, 11)  #10格
plt.bar(index[0::2], points, label="Points")  # x軸用index有10格，index[0::2]表示從第一個開始跳兩格，也就是1,3,5,7,9
plt.bar(index[1::2], rebounds, label="Rebounds") # y軸用index有10格，index[1::2]表示從第一個開始跳兩格，也就是2,4,6,8,10
plt.xticks(index[0::2], points.index)  #在x軸的1,3,5,7,9位置放索引(C,PF,PG....)
plt.legend()
plt.ylabel("Points and Rebounds")
plt.xlabel("Position")
plt.title("NBA Houston Rockets") 
plt.show()
```

# 使用多個圖形和軸
matplotlib的Figure圖形物件是一個容器，可以同時繪出多張圖表，而每一張圖表是在繪在指定的軸(Axes)上，這就是子圖表(subplots)
使用subplots()函數可以在指定儲存格繪製子圖表

```python
import numpy as np

def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t)

t1 = np.arange(0.0, 5.0, 0.1)
t2 = np.arange(0.0, 5.0, 0.02)

plt.figure() 
plt.subplot(211)
plt.plot(t1, f(t1), 'bo', 
         t2, f(t2), 'k')

plt.subplot(212)
plt.plot(t2, np.cos(2*np.pi*t2), 'r--')
plt.show()
```

```python
#Ch13_4_1
import matplotlib.pyplot as plt
import math

x = [0,0.5,1,1.5,2,2.5,3,3.5,4,4.5,5,
     5.5,6,6.5,7,7.5,8,8.5,9,9.5,10]
sinus = [math.sin(v) for v in x]
cosinus = [math.cos(v) for v in x]
plt.subplot(2, 1, 1)  #圖的排列(列,欄,順序):此例是2*1表格中的第一位，順序是從左到右，再從上到下
plt.plot(x, sinus, "r-o")
plt.subplot(2, 1, 2)
plt.plot(x, cosinus, "g--")
plt.show()

```

# 繪製雙Y軸圖

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(0., np.e, 0.01)
y1 = np.exp(-x)
y2 = np.log(x)

fig = plt.figure()

ax1 = fig.add_subplot(111)
ax1.plot(x, y1)
ax1.set_ylabel('Y values for exp(-x)')
ax1.set_title("Double Y axis")

# 使用twinx是添加y轴的坐标轴
# 使用twiny是添加x轴的坐标轴
ax2 = ax1.twinx()  # this is the important function
ax2.plot(x, y2, 'r')
ax2.set_xlim([0, np.e])
ax2.set_ylabel('Y values for ln(x)')
ax2.set_xlabel('Same X for both exp(-x) and ln(x)')

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(0,10,0.1)
y1 = 0.05 * x ** 2
y2 = -1 * y1

fig,ax1 = plt.subplots()

# 使用twinx是添加y軸的座標軸
# 使用twiny是添加x軸的座標軸
ax2 = ax1.twinx()  
ax1.plot(x,y1,'g-')
ax2.plot(x,y2,'b-')

ax1.set_xlabel('X data')
ax1.set_ylabel('Y1 data',color = 'g')
ax2.set_ylabel('Y2 data',color = 'b')

plt.show()
```

### 圖中圖

```python
import matplotlib.pyplot as plt

fig = plt.figure()
x = [1,2,3,4,5,6,7]
y = [1,3,4,2,5,8,6]

#below are all percentage
left, bottom, width, height = 0.1, 0.1, 0.8, 0.8
#使用plt.figure()顯示的是一個空的figure
#如果使用fig.add_axes會添加軸
ax1 = fig.add_axes([left, bottom, width, height])# main axes
ax1.plot(x,y,'r')
ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax1.set_title('title')

ax2 = fig.add_axes([0.2, 0.6, 0.25, 0.25])  # inside axes
ax2.plot(y, x, 'b')
ax2.set_xlabel('x')
ax2.set_ylabel('y')
ax2.set_title('title inside 1')

# different method to add axes
plt.axes([0.6, 0.2, 0.25, 0.25])
plt.plot(y[::-1], x, 'g')
plt.xlabel('x')
plt.ylabel('y')
plt.title('title inside 2')

plt.show()
```

### animation動畫

```python
import numpy as np
from matplotlib import pyplot as plt
from matplotlib import animation

fig,ax = plt.subplots()

x = np.arange(0,2*np.pi,0.01)
#因为这里返回的是一个列表，但是我们只想要第一个值
#所以这里需要加,号
line, = ax.plot(x,np.sin(x))

def animate(i):
    line.set_ydata(np.sin(x + i/10.0))#updata the data
    return line,

def init():
    line.set_ydata(np.sin(x))
    return line,


# call the animator.  blit=True means only re-draw the parts that have changed.
# blit=True dose not work on Mac, set blit=False
# interval= update frequency
#frames帧数
ani = animation.FuncAnimation(fig=fig, func=animate, frames=100, init_func=init,
                              interval=20, blit=False)

plt.show()
```

### 3D圖

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = Axes3D(fig)
#X Y value
X = np.arange(-4,4,0.25)
Y = np.arange(-4,4,0.25)
X,Y = np.meshgrid(X,Y)
R = np.sqrt(X**2 + Y**2)
#hight value
Z = np.sin(R)

ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap=plt.get_cmap('rainbow'))
"""
============= ================================================
        Argument      Description
        ============= ================================================
        *X*, *Y*, *Z* Data values as 2D arrays
        *rstride*     Array row stride (step size), defaults to 10
        *cstride*     Array column stride (step size), defaults to 10
        *color*       Color of the surface patches
        *cmap*        A colormap for the surface patches.
        *facecolors*  Face colors for the individual patches
        *norm*        An instance of Normalize to map values to colors
        *vmin*        Minimum value to map
        *vmax*        Maximum value to map
        *shade*       Whether to shade the facecolors
        ============= ================================================
"""

# I think this is different from plt12_contours
ax.contourf(X, Y, Z, zdir='z', offset=-2, cmap=plt.get_cmap('rainbow'))
"""
==========  ================================================
        Argument    Description
        ==========  ================================================
        *X*, *Y*,   Data values as numpy.arrays
        *Z*
        *zdir*      The direction to use: x, y or z (default)
        *offset*    If specified plot a projection of the filled contour
                    on this position in plane normal to zdir
        ==========  ================================================
"""
ax.set_zlim(-2, 2)
plt.show()
```
